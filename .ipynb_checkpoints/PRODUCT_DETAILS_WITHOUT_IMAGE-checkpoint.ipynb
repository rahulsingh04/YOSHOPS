```python
import requests
from bs4 import BeautifulSoup
from urllib.request import Request
from urllib.request import urlopen
import pandas as pd
from autoscraper import AutoScraper

url = "https://yoshops.com/products?page="


def search(text, sub1, sub2, buffer1, buffer2):
    index1 = t.index(sub1)
    index2 = t.index(sub2)
    val = text[index1 + len(sub1) + buffer1: index2 - buffer2]
    return val

yoshops_list = []
product_name_list = []
product_price_list = []
product_hyp_link_list = []
for i in range(1, 12):
    new_url = url + str(i)
    req = Request(new_url, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    soup = BeautifulSoup(webpage, 'html.parser')
    table = soup.find('div', class_='products')
    lists = table.find_all('div', class_="col-sm-3 col-xs-6")
    data = []
    d = []
    d1 = []
    hyp_link = []
    for row in lists:
        d.append(row)
        t = str(row)
        l = "//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png"
        if l in t:
            hyp1 = search(t, "href=", '<div class="product-thumb">',1,1)
            hyp = hyp1[:-2]
            d1.append(hyp)
    for link in d1:
        url_1 = "https://yoshops.com"
        new_url_1 = url_1 + str(link)
        hyp_link.append(new_url_1)

    print("hyperlink size", len(hyp_link))
    #product_hyp_link_list.append(hyp_link)

    for products_link in hyp_link:
        product_hyp_link_list.append(products_link)
        req_2 = Request(products_link, headers={'User-Agent': 'Mozilla/5.0'})
        webpage_2 = urlopen(req_2).read()
        soup_2 = BeautifulSoup(webpage_2, 'html.parser')
        table_2 = soup_2.find('div', class_='content')
        product_details = table_2.find_all('div', {"id": "product-description-height"})
        ##print("productdetails- ", product_details)
        t = str(product_details)
        products_name = search(t, "Buy", "at  price Rs.",0,0)
        products_price = search(t, "price Rs.", " from Yoshops.com",0,0)
        product_name_list.append(products_name)
        product_price_list.append(products_price)
yoshops_list_page = {'Product Name': product_name_list,
                     'Product Price (Rupees)': product_price_list,
                     'Product Link': product_hyp_link_list
                    }
yoshops_list_without_img = pd.DataFrame(yoshops_list_page)
#yoshops_list.append(yoshops_list_without_img)
yoshops_list_without_img
#print(yoshops_list)
```
