```python
from selenium.webdriver.chrome.service import Service
import time
s=Service("C:\\BrowserDriver\\chromedriver_win32\\chromedriver.exe")
browser = webdriver.Chrome(service=s)
```

## One can use find_elements_by_xpath
Take a good look at the format of xpath.

```python
#### PRODUCT
```

```python
url="https://yoshops.com/products?page=1" 
browser.get(url)
time.sleep(5)
productName = browser.find_elements_by_xpath("//span[@class='product-title']")
```

```python
productName[0].text
```

### Product ratings


```python
time.sleep(5)
productRating = browser.find_elements_by_xpath("//span[@class='sr-only']")
```

```python
productRating[1].text
```

### Product reviews


```python
url='https://yoshops.com/products/reliance-jio-phone1'
browser.get(url)
time.sleep(5)
productReviews = browser.find_elements_by_xpath("//div[@class='content-review']")
totalReviews=len(productReviews)
```
