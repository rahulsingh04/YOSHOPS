### Web scraping in Python with BeautifulSoup


```python
# import necessary libraries
import urllib
from bs4 import BeautifulSoup
```

```python
### Get access to the HTML of the Page:l
url = "https://yoshops.com/products?keywords=river+prawn"
html = urllib.request.urlopen(url)
```

```python
# The HTML content of the page is in html. Transform it into a BeautifulSoup object. 
soup = BeautifulSoup(html,'html.parser')
```

### BeautifulSoup provides find() and find_all() methods for finding anything on the webpage.
##### Let's explore them.

#### The find method finds out the first tag with the specified name or id.
### Lets do for 'span' tag with class 'product-title' which corresponds to the product "River Prawn Small - 4cm 250 ### gm (Berhampur)" in the above figure.

```python
### lets look for the div tags.
span_tag = soup.find('span',  {"class":"product-title"})
len(span_tag)
```

```python
### lets look for the div tags.
span_tag1 = soup.find_all("span", {"class" : "product-title"})
len(span_tag1)
```

###  find method found only the first 'span' tag with class 'product-title' and hence the length is 1.
###  Let us look at what span_tag hold

### Similarly, go for extracting other information.
### Suppose you wish to extract the product price.


```python
div_tags = soup.findAll('div',{'class':"product-price"})
```

```python
original_product_prices=[]
for i in div_tags:
    product_price=i.findAll('span')[0].text
    original_product_prices.append(product_price)
```

```python
original_product_prices
```
